# Generating XGBOOST Model...

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import xgboost as xgb
```

```python
checkpoint5_traffic = pd.read_csv('C:/Users/Admin/Documents/GitHub/CZ4041-NYC-Taxi-Fare/checkpoint5_SGDR_traffic.csv')
checkpoint5_traffic
```

```python
labels = checkpoint5_traffic[['fare_amount']]
labels
```

```python
del checkpoint5_traffic['fare_amount']
checkpoint5_traffic
```

```python
del checkpoint5_traffic['time']
checkpoint5_traffic
```

```python
params = {
    "max_depth": 7,
    "subsample": 0.9,
    "eta": 0.03,
    "colsample_bytree": 0.9,
    "random_state": 2666,
    "objective": "reg:linear",
    "eval_metric": "rmse",
    "silent": 1
}
```

```python
print("splitting...")
#splitting data into train test 
x_train, x_test, y_train, y_test = train_test_split(checkpoint5_traffic, labels, test_size=0.2, random_state=42)
```

```python
def XGBmodel(x_train, x_test, y_train, y_test, params):
    matrix_train = xgb.DMatrix(x_train, label=y_train)
    matrix_test = xgb.DMatrix(x_test, label=y_test)
    
    model = xgb.train(params=params,
                      dtrain=matrix_train,num_boost_round=5000, 
                      early_stopping_rounds=10,evals=[(matrix_test,'test')])
    return model

#start_time = time.time()
model = XGBmodel(x_train, x_test, y_train, y_test, params)

model.save_model("xgboost_traffic_LR1.json")
model.save_model("xgboost_traffic_LR1.txt")
```
